```python
from sympy import *
init_printing(use_latex=True)

from sympy.physics.mechanics import *
```

```python
# Definimos nuestra propia clase para que los versores sean IJK
# Por Juan Luis Cano 
class IJKReferenceFrame(ReferenceFrame):
    def __init__(self, name):
        super().__init__(name, latexs=['\mathbf{%s}_{%s}' % (idx, name) for idx in ("i", "j", "k")])
```

En primer lugar se crean los sistemas de referencia:

* Tierra (0)
* Arbol (A)
* Auxiliar 1 (A1)

```python
Tierra = IJKReferenceFrame('0')
Arbol = IJKReferenceFrame('A')
A1 = IJKReferenceFrame('A1')
```

Se orienta el sistema $A$ con respecto al $0$ y $A1$ con respecto al sistema $A$

```python
Arbol.orient(Tierra, 'body', (pi,0,pi), 'zyx')

psi = dynamicsymbols('\psi')
A1.orient(Arbol, 'body', (psi, 0, 0), 'zyx')
```

```python
Arbol.dcm(Tierra)
```

```python
Arbol.dcm(A1)
```

Una vez orientados los sistemas de referencia se puede obtener la matriz de cosenos directores que cambia de uno a otro, pero ni siquiera la vamos a necesitar.

Se define el punto $A$. Los puntos no se definen ligados a sistemas de referencia, pero sí sus velocidades. Esto resulta un poco extraño...

```python
A = Point('A')

V = symbols('V')
A.set_vel(Tierra, V*Tierra.x)
A.set_vel(Arbol, 0)

#la velocidad del punto A respecto a los ejes tierra es:
A.vel(Tierra)
```

```python
# y si la expresamos en ejes arbol:
A.vel(Tierra).express(Arbol)
```

Se define ahora un punto $P$ a distancia $r$ sobre la pala:

```python
r = symbols('r')
P = A.locatenew('P', r*A1.x)
```

Conocida la velocidad de $A$ y la posición de $P$ con respecto a $A$, sabiendo que están en el mismo sistema de referencia ($A1$) se puede obtener la velocidad de $P$:

```python
#velocidad del punto P en ejes tierra
vel_P_0 = P.v2pt_theory(A, Tierra, A1)
vel_P_0
```

```python
#se puede expresar en otros ejes
vel_P_0.express(A1)
```

```python
omega = symbols('\Omega')
vel_P_0 = vel_P_0.express(A1).subs(diff(psi), omega)
vel_P_0
```

```python
#velocidad del punto P en ejes Arbol
vel_P_A = P.v2pt_theory(A, Arbol, A1)
vel_P_A
```

```python
vel_P_A.express(Arbol)
```

La componente tangencial de la velocidad es:

```python
V_T  = vel_P_0.dot(A1.y)
V_T
```

```python
solucion_r = solve(V_T, r)
solucion_r
```

Esta expresión se puede particulariza utilizando `.subs()` o transformarla con `simpify` para obtener un mejor rendimiento

```python
for ii in range(1,6):
    radio = solucion_r[0].subs([(V,100), (omega, 12), (psi,ii*(2*pi/5))])
    print(radio)
```

Para representar el círculo de inversión:

```python
sol_particular = solucion_r[0].subs([(V,100), (omega, 12)])

#la función plot no acepta pintar en función de psi porque
#es una función de t. Hay que transformarlo en un símbolo primero

x = symbols('x')
sol_particular = sol_particular.subs(psi, x)

%matplotlib inline

plot(sol_particular, (x, 0, 2*pi))
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# sympy todavía no está bien preparado para pintar en polares
# se podría hacer usando plot_parametrics
# pero se ha usado matplotlib

xx = np.linspace(0,2*np.pi,100)
y = lambdify(x, sol_particular, 'numpy')
yy = y(xx)

plt.polar(xx, yy, lw=2)
```

Si se quieren pintar varios casos de velocidad:

```python
velocidades = np.linspace(0.1,100,6)
solucion = solucion_r[0].subs([(omega, 12), (psi, x)])
plt.figure()

for ii in velocidades:
    solucion_caso = solucion.subs(V,ii)
    solucion_caso_num = lambdify(x, solucion_caso, 'numpy')
    
    xx = np.linspace(0,2*np.pi,100)
    yy = solucion_caso_num(xx)
    
    plt.polar(xx, yy, lw=2, color='b', alpha=ii/100)
```

## Deslizadores

```python
def circulo_inversion(OMEGA, VEL):
    solucion = solucion_r[0].subs([(omega, OMEGA), (psi, x)])
    plt.figure()

    solucion_caso = solucion.subs(V,VEL)
    solucion_caso_num = lambdify(x, solucion_caso, 'numpy')

    xx = np.linspace(0,2*np.pi,150)
    yy = solucion_caso_num(xx)

    fig, ax = plt.subplots(1,1, subplot_kw=dict(polar=True))
    
    fig.set_size_inches((6,6))
    ax.plot(xx, yy, lw=2, color='b', alpha=ii/100)
    ax.fill_between(xx, yy, alpha=0.5, color='gray')
    
    ax.set_rmax(15)
```
